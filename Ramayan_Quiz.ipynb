```python
'''

Quiz Mini Project

- Date         : 25/09/2024


'''


'''
- Date         : 27/09/2024

Adding additional features to Python Mini Project on a Quiz game


'''


import time

# Defining a list of questions with difficulty levels

questions = {

    "easy": [
        {
            "question": " Who is the author of the Ramayan?",
            "options": ["A) Valmiki", "B) Tulsidas", "C) Vyasa", "D) Kalidasa"],
            "answer": "A"
        },

        {
            "question": " Who is the main character of the Ramayan?",
            "options": ["A) Lakshmana", "B) Ravana", "C) Ram", "D) Bharata"],
            "answer": "C"
        },

        {
            "question": " Who was Ram's father?",
            "options": ["A) Ravana", "B) Dasharatha", "C) Bharata", "D) Janaka"],
            "answer": "B"
        },

        {
            "question": " What was the name of Ram’s kingdom?",
            "options": ["A) Ayodhya", "B) Hastinapura", "C) Lanka", "D) Mathura"],
            "answer": "A"
        },

        {
            "question": " Who is the mother of Ram?",
            "options": ["A) Kaikeyi", "B) Kaushalya", "C) Sumitra", "D) Mandodari"],
            "answer": "B"
        },

        {
            "question": " Who is the twin brother of Lakshman?",
            "options": ["A) Bharata", "B) Vibhishana", "C) Shatrughna", "D) Rama"],
            "answer": "C"
        },

        {
            "question": " Who is the mother of Lakshman and Shatrughna?",
            "options": ["A) Sumitra", "B)  Kaikeyi", "C) Kaushalya", "D) Tara"],
            "answer": "A"
        },

        {
            "question": " Who asked Dasharatha to send Ram to help with his yagna",
            "options": ["A) Vashishta", "B) Vishwamitra", "C) Agastya", "D) Parashurama"],
            "answer": "B"
        },

        {
            "question": " What is the name of Ram's wife?",
            "options": ["A) Mandavi", "B) Mandodari", "C) Urmila", "D) Sita"],
            "answer": "D"
        },

        {
            "question": " Who is the mother of Bharata?",
            "options": ["A) Kaushalya", "B) Sumitra", "C) Kaikeyi", "D) Sita"],
            "answer": "C"
        },

       {
            "question": " Who was Sita's father?",
            "options": ["A) Janaka", "B) Dasharatha", "C) Bharata", "D) Ravana"],
            "answer": "A"
        },

         {
            "question": " What was the name of the kingdom ruled by King Janaka?",
            "options": ["A) Ayodhya", "B) Mithila", "C) Lanka", "D) Mathura"],
            "answer": "B"
        },

        {
            "question": " Who poisoned the mind of Kaikeyi?",
            "options": ["A) Shurpanakha", "B) Kausalya", "C) Sumitra", "D) Manthara"],
            "answer": "D"
        },


        {
            "question": " What was the reason for Ram's exile to the forest?",
            "options": ["A) To hunt deer", "B) A promise made by King Dasharatha", "C) To find Sita", "D) To defeat Ravana"],
            "answer": "B"
        },

        {
            "question": " Who helped Sita, Ram, and Lakshman to cross the Sarayu River?",
            "options": ["A) Nala", "B) Guha", "C) Hanuman", "D) Varuna"],
            "answer": "B"
        },


        {
            "question": " Which demoness tried to mislead Ram and Lakshman in the forest?",
            "options": ["A) Surpanakha", "B) Tataka", "C) Trijata", "D) Sulochana"],
            "answer": "A"
        },

       {
            "question": " Who kidnapped Sita?",
            "options": ["A) Vali", "B) Indrajit", "C) Ravan", "D) Maricha"],
            "answer": "C"
        },

        {
            "question": " Which bird tried to save Sita when she was being abducted by Ravana?",
            "options": ["A) Jatayu", "B) Garuda", "C) Sampati", "D) Shuka"],
            "answer": "A"
        },

       {
            "question": " What was the name of Ravana’s flying chariot?",
            "options": ["A) Agni Rath", "B) Garuda", "C) Pushpaka Vimana", "D) Indra Vimana"],
            "answer": "C"
        },


         {
            "question": " To whom did Lord Rama give his ring as a token to present to Sita?",
            "options": ["A) Sugriva", "B) Hanuman", "C) Jambavan", "D) Angada"],
            "answer": "B"
        }

    ],

    "medium": [

        {
            "question": " What is the name of Bharata's wife?",
            "options": ["A) Urmila", "B) Mandavi", "C) Shrutakirti", "D) Sita"],
            "answer": "B"
        },

        {
            "question": " What is the name of Lakshmana's wife?",
            "options": ["A) Sita", "B) Mandavi", "C) Urmila", "D) Shrutakirti"],
            "answer": "C"
        },

        {
            "question": " Which name of Sita is derived from her father's name?",
            "options": ["A) Vaidehi", "B) Maithili", "C) Bhumija", "D) Janaki"],
            "answer": "D"
        },

        {
            "question": " Who is the twin brother of Sugriv (the monkey king)?",
            "options": ["A) Vali", "B) Angad", "C) Jambavan", "D) Nala"],
            "answer": "A"
        },

       {
            "question": " What is the name of Shatrughna's wife?",
            "options": ["A) Shrutakirti", "B) Mandavi", "C) Urmila", "D) Sita"],
            "answer": "A"
        },

        {
            "question": " Who is the daughter of King Janaka?",
            "options": ["A) Shrutakirti", "B) Mandavi", "C) Urmila", "D) Anjana"],
            "answer": "C"
        },


        {
            "question": " Who is the mother of Sita and Urmila?",
            "options": ["A) Sunaina", "B) Kaushalya", "C) Sumitra", "D) Mandodari"],
            "answer": "A"
        },

        {
            "question": " Who is the guru of Ram?",
            "options": ["A) Vishwamitra", "B) Parashurama", "C) Vashishta", "D) Agastya"],
            "answer": "C"
        },

       {
            "question": " What was the condition in Sita's swayamvar?",
            "options": ["A) To lift the divine bow", "B) To break Shiva dhanush", "C) To defeat a demon",
                        "D) To string Shiva dhanush"],
            "answer": "D"
        },

        {
            "question": " What is the name of the monkey king who helped Rama?",
            "options": ["A) Sugriva", "B) Hanuman", "C) Vali", "D) Angada"],
            "answer": "A"
        },

        {
            "question": " Who is the mother of Hanuman?",
            "options": ["A) Anjana", "B) Swayamprabha", "C) Tara", "D) Ruma"],
            "answer": "A"
        },

        {
            "question": " Who is the father of Hanuman?",
            "options": ["A) Janaka", "B) Kesari", "C) Surya", "D) Vali"],
            "answer": "B"
        },

       {
            "question": " Who is the guru of Hanuman?",
            "options": ["A) Brihaspati", "B) Surya Dev", "C) Vishwamitra", "D) Vashishta"],
            "answer": "B"
        },

        {
            "question": " Why is Hanuman staying with the monkey king Sugriv?",
            "options": ["A) To protect Sugriv from Bali", "B) As a punishment", "C) To search for Sita",
                        "D) To fulfill a promise made to his guru Surya Dev"],
            "answer": "D"
        },

       {
            "question": " Who is the father of Angada?",
            "options": ["A) Sugriv", "B) Hanuman", "C) Vali", "D) Jambavan"],
            "answer": "C"
        },

        {
            "question": " Who helped Ram to cross the ocean in order to reach Lanka?",
            "options": ["A) Nala & Nila", "B) Vibhishana", "C) Hanuman", "D) Varuna"],
            "answer": "A"
        },

        {
            "question": " What was the name of the bridge built to cross to Lanka?",
            "options": ["A) Hanuman Setu", "B) Lakshman Setu", "C) Varuna Setu", "D) Ram Setu"],
            "answer": "D"
        },

        {
            "question": " Who became the king of Lanka after Ravana’s death?",
            "options": ["A) Kumbhakarna", "B) Vibhishana", "C) Indrajit", "D) Ram"],
            "answer": "B"
        },

        {
            "question": " Who are the sons of Ram?",
            "options": ["A) Lava and Kusha", "B) Bharata and Shatrughna", "C) Angad and Chandraketu",
                        "D) Indrajit and Akshayakumara"],
            "answer": "A"
        },

       {
            "question": " Who is the guru of the sons of Ram?",
            "options": ["A) Vashishta", "B) Vishwamitra", "C) Valmiki", "D) Agastya"],
            "answer": "C"
        }

    ],

    "hard": [

        {
            "question": " Who is the father of Mandavi and Shrutakirti?",
            "options": ["A) King Janaka", "B) King Dasharatha", "C) King Kushadhwaja", "D) King Ashwapati"],
            "answer": "C"
        },

        {
            "question": " Who is the mother of Mandavi and Shrutakirti?",
            "options": ["A) Sunaina", "B) Chandrabhaga", "C) Kaushalya", "D) Sumitra"],
            "answer": "B"
        },

       {
            "question": " Who was the king of Lanka before Ravana?",
            "options": ["A) Surya", "B) Mandodari's father", "C) Indra", "D) Kubera"],
            "answer": "D"
        },

       {
            "question": " Why did Sita get the name Bhumija?",
            "options": ["A) As she was born from the fire", "B) As she was born from the earth",
                        "C) As she was the daughter of King Janaka", "D) As she was a goddess"],
            "answer": "B"
        },

        {
            "question": " Who is the wife of Vali?",
            "options": ["A) Tara", "B) Ruma", "C) Mandavi", "D) Anjana"],
            "answer": "A"
        },

        {
            "question": "  What was the boon granted to Vali?",
            "options": ["A) Immortality", "B) make him unconquerable to all the creations of Brahma, except for humans",
                        "C) Ability to change forms", "D) Ability to absorb half of his opponent's strength"],
            "answer": "D"
        },


        {
            "question": " Who is the wife of Sugriva?",
            "options": ["A) Tara", "B) Ruma", "C) Mandavi", "D) Anjana"],
            "answer": "B"
        },

       {
            "question": " Who took care of Sita in Lanka?",
            "options": ["A) Mandodari", "B) Surpanakha", "C) Trijata", "D) Sulochana"],
            "answer": "C"
        },

       {
            "question": " Who is the father of Trijata?",
            "options": ["A) Vibhishana", "B) Ravana", "C) Meghanadha", "D) Kumbhakarna"],
            "answer": "A"
        },

       {
            "question": " Where did Hanuman find Sita in Lanka?",
            "options": ["A) Ravana’s palace", "B) In a temple", "C) In the forest", "D) Ashok Vanam"],
            "answer": "D"
        },

       {
            "question": "  Who is the father of Indrajith?",
            "options": ["A) Kumbhakarna", "B) Ravana", "C) Vibhishana", "D) Akshayakumara"],
            "answer": "B"
        },

        {
            "question": "  What is the birth name of Indrajith?",
            "options": ["A) Kumbhakarna", "B) Ravana", "C) Meghanadha", "D) Akshayakumara"],
            "answer": "C"
        },

       {
            "question": "  Why did Meghanadhan get the name Indrajith?",
            "options": ["A) As he defeated Indra Dev", "B) As he was born to Ravana",
                        "C) As he conquered the gods", "D) As he married to Naag kanni"],
            "answer": "A"
        },

       {
            "question": "  Who is the wife of Indrajith?",
            "options": ["A) Tara", "B) Trijata", "C) Ruma", "D) Sulochana"],
            "answer": "D"
        },

        {
            "question": "  How did Hanuman warn Ravana before leaving Lanka?",
            "options": ["A) By threatening him", "B) By burning half of Lanka",
                        "C) By sending a message", "D) By capturing his soldiers"],
            "answer": "B"
        },

        {
            "question": "  Who challenged the entire Ravana's court to move his leg?",
            "options": ["A) Hanuman", "B) Sugriv", "C) Angada", "D) Vibhishana"],
            "answer": "C"
        },

        {
            "question": "  What boon did Lord Brahma grant to Ravana?",
            "options": ["A) Immortality", "B) He should not be killed by any of Brahma's creations, except humans",
                        "C) Ability to change forms", "D) Control over animalsr"],
            "answer": "B"
        },


       {
            "question": "  Who killed Indrajith?",
            "options": ["A) Lakshmana", "B) Rama", "C) Hanuman", "D) Angada"],
            "answer": "A"
        },


        {
            "question": "  Through which disciples did Valmiki start to spread the Ramayana in this world?",
            "options": ["A) Sages", "B) Monkeys", "C) Sons of Ram", "D) The people of Ayodhya"],
            "answer": "C"
        },


        {
            "question": "  Who narrated the entire Ramayana to Lord Rama and the people of Ayodhya?",
            "options": ["A) Lava and Kusha", "B) Bharata and Shatrughna", "C) Angad and Chandraketu",
                        "D) Indrajit and Akshayakumara"],
            "answer": "A"
        }

    ]
}

# Global leaderboard
leaderboard = [
    {'nickname': 'Ballu', 'score': 20, 'stars': 5},
    {'nickname': 'Subbu', 'score': 18, 'stars': 4},
    {'nickname': 'Jaggu', 'score': 15, 'stars': 4},
    {'nickname': 'Chotu', 'score': 12, 'stars': 3},
    {'nickname': 'Babloo', 'score': 10, 'stars': 2}
]

# Define the ask_question function
def ask_question(question, options, correct_answer, time_limit):

    # Ask the question
    print(question)
    # Display all options
    for option in options:
        print(option)

    # Start the timer
    start_time = time.time()
    user_answer = input("\nYour answer (A/B/C/D) or 'E' to exit:\n\n").strip().upper()

    # Early exit
    if user_answer == "E":
        print("\nExiting the quiz.\n")
        return -1, 0  # Return -1 for early exit and 0 as the elapsed time

    # Calculate elapsed time
    elapsed_time = time.time() - start_time
    if elapsed_time > time_limit:  # Timeout
        print("\nTime's up! You didn't answer in time.\n")
        print(f"**************************************************************************\n")
        return 0, elapsed_time  # Return 0 for time out and the time taken

    if user_answer == correct_answer:
        print("\nCorrect!\n")
        print(f"**************************************************************************\n")
        return 1, elapsed_time  # Return score as 1 (incase of correct answer) and the time taken

    else:
        print(f"\nWrong! The correct answer was {correct_answer}.\n")
        print(f"**************************************************************************\n")
        return 0, elapsed_time  # Return 0 for incorrect answer and the time taken

# Display user instructions (additional feature)
def display_instructions():

    instructions = [
        "You will be presented with multiple-choice questions (A/B/C/D).",
        "You have a limited amount of time to answer each question.",
        "You need to choose the difficulty level (easy/medium/hard), based on which the questions will be asked.",
        "Time limit for easy level is 10 seconds.",
        "Time limit for medium level is 15 seconds.",
        "Time limit for hard level is 20 seconds.",
        "Stars will be given to those who answer the questions very fast. There are 5 stars in total!",
        "If you answer four questions in under 5 seconds for easy level, you will get a star.",
        "If you answer four questions in under 7 seconds for medium level, you will get a star.",
        "If you answer four questions in under 10 seconds for hard level, you will get a star."
    ]

    print("\n\t\t\t-------Instructions-------\n")
    for i in range(0, len(instructions), 2):
        print(instructions[i])
        print()
        if i + 1 < len(instructions):
            print(instructions[i + 1])
            print()

        # Choice to skip or continue the instructions
        user_input = input("Enter 'c' to continue or 's' to skip instructions:\n").strip().lower()

        if user_input == 's':
            print("\nSkipping instructions...\n")
            break
        elif user_input != 'c':
            print("\nInvalid input! Please press 'c' to continue or 's' to skip.\n")
            continue

# Define the quiz function
def quiz(nickname):
    global leaderboard
    print(f"**************************************************************************\n")
    print(f"\t\t\t\tRamayan\n")
    print(f"**************************************************************************\n")
    print("\n\t\t\t       Jai SiyaRam!!\n")

    # Display user instructions
    display_instructions()

    # Get the choice for difficulty level
    difficulty = input("\nChoose the difficulty level (Easy/Medium/Hard):\n").strip().lower()
    if difficulty not in questions:
        print("Invalid option. Defaulting to easy level.\n")
        difficulty = "easy"
    print()

    score = 0
    total_stars = 0
    time_limit = 10
    questions_answered = 0
    time_taken_for_four = 0

    # Time limits based on difficulty level (additional feature)
    if difficulty == 'easy':
        time_limit = 10
        star_time_limit = 5
    elif difficulty == 'medium':
        time_limit = 15
        star_time_limit = 7
    else:  # incase of hard
        time_limit = 20
        star_time_limit = 10

    # Ask questions based on difficulty level
    for q in questions[difficulty]:
        result, elapsed_time = ask_question(q["question"], q["options"], q["answer"], time_limit)

        if result == -1:  # Early exit
            break

        score += result
        questions_answered += 1
        time_taken_for_four += elapsed_time

        # Check if user answered 4 questions and earned a star (additional feature)
        if questions_answered % 4 == 0:
            if time_taken_for_four <= star_time_limit * 4:
                total_stars += 1
                print(f"\n\t\t\t  ⭐ You've earned a star!\n")
                print(f"**************************************************************************\n")
            time_taken_for_four = 0

        # Motivational messages based on score (additional feature)
        if score > 0 and score % 5 == 0:
            if score == 5:
                print("\n\t\t\t  You rock! Keep it up!\n")
                print(f"**************************************************************************\n")
            elif score == 10:
                print("\n\t\t\t  You're doing great!\n")
                print(f"**************************************************************************\n")
            elif score == 15:
                print("\n\t\t\t  Fantastic job! You're on fire!\n")
                print(f"**************************************************************************\n")
            elif score == 20:
                print("\n\t\t\t  You're awesome! Keep rocking!\n")
                print(f"**************************************************************************\n")


    # Displaying the stars earned (additional feature)
    stars = []
    for i in range(total_stars):
        stars.append("⭐")

    # Add to leaderboard
    leaderboard.append({'nickname': nickname, 'score': score, 'stars': total_stars})
    leaderboard.sort(key=lambda x: x['score'], reverse=True)

    # Display final score
    if result == -1:
        print(f"**************************************************************************\n")
        print(f"\t\t\tYour final score is: {score}/{len(questions[difficulty])}\n")
        print(f"\n\t\t\tYou earned {total_stars} out of 5 stars!\n")
        if total_stars:
            print("\n\t\t\t    ", " ".join(stars))
            print()
            print()
        print(f"**************************************************************************\n")
    else:
        print(f"\n\t\t\tYour final score is: {score}/{len(questions[difficulty])}\n")
        print(f"\n\t\t\tYou earned {total_stars} out of 5 stars!\n")
        print("\n\t\t\t    ", " ".join(stars))
        print()
        print()
        print(f"**************************************************************************\n")

    # Display leaderboard
    print("\n\t\t\t--------Leaderboard--------\n")
    i = 0
    while i < min(5, len(leaderboard)):
        player = leaderboard[i]
        print(f"\n\t\t {i+1}. {player['nickname']} -> Score: {player['score']}, Stars: {'⭐' * player['stars']}")
        i += 1
    print(f"\n**************************************************************************\n")


def main():
    nickname = input("\nEnter your nickname: ").strip()
    while True:
        quiz(nickname)

        # Prompt the user if they want to replay
        play_again = input("\nDo you want to play again? (yes/no):\n").strip().lower()

        # To exit the quiz
        if play_again != 'yes':
            print("\n\tThanks for playing!\n")
            print("\n\t  Jai SiyaRam!!\n")
            break


if __name__ == "__main__":
    main()
```
